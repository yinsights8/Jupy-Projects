```python
# Hangman :- save your man
import random
from IPython.display import clear_output
words = ["yash","kunal","potato","maggi","nuddels"]
display = []
lives = 6
random_word = random.choice(words)

for _ in range(len(random_word)):
    display += "_"
#print(display)

print(f"Random_word is {random_word}")

should_continue = True
while should_continue:

    guessed_word = input("Guess a  Word: ")
    clear_output()
    if guessed_word in display:
        print(f"You already guessed a word: {guessed_word}")

    for position in range(len(random_word)):
        letter = random_word[position]
        if guessed_word == letter:
            display[position] = letter
    
    if guessed_word not in random_word:
        lives -= 1
        print(f"Wrong word. You have {lives} life of 6\n")
        if lives == 0:
            should_continue = False
            print(f"The WORD was '{random_word.upper()}'\n")
            print("Died ü§™\n")
            print("YOU LOSE")
    
    if "_" not in display:
        should_continue = False
        print("i am SAVED‚Ä¶ üòÆ‚Äçüí® ‚Ä¶")
        print("You WIN")
        
    print(" ".join(display))
    print()
    print(stages[lives])
```
